```python
import os
import bs4
import requests
```

```python
def http_folder_download(url, folder_name):
    r = requests.get(url)
    # TO BE ADDED: assert r.status_code = 200
    html = bs4.BeautifulSoup(r.text, "html.parser")
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    for line in html.find_all("a"):
        filename = line["href"]

        if filename != '../':
            print("writing file: " + filename)
            r = requests.get(url + line["href"])
            with open (folder_name + '/' + filename, 'wb') as f:            
                f.write(r.content)
                f.close()

    print("Done downloading: " + url)
```

```python
huc_list = ['101702', '102300', '102400', '102801',
            '102802', '071100', '071000', '070801', 
            '070802', '070600']

for huc in huc_list:
    url = 'https://web.corral.tacc.utexas.edu/nfiedata/HAND/{}/'.format(huc)
    folder_name = huc
    print(folder_name, url)

    http_folder_download(url, folder_name)
```
